```python
df20 = pd.DataFrame([['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long']],columns=['color','size','sex','length'])
df20
```

```python
# sklearn LabelEncoder ()   (ord())
from sklearn.preprocessing import LabelEncoder
class_label = LabelEncoder()                     # LabelEncoder() class_labellabel inverse
df20['size_label'] = class_label.fit_transform(df20['size'].values)  # size LabelEncoder  (python ord())
df20
```

```python
# inverse_transform label 
df20['size_label_inverse'] = class_label.inverse_transform(df20['size_label'])
df20
```

```python
# LabelEncoder() ord(), ASCII code, ()map 
df20["size_label_map"]=df20["size"].map({'M': 0, 'L': 1,'XL':2})
df20
```

```python
# Categorical() category => category ( ['M' < 'L' < 'XL'])
category = pd.Categorical(df20['size'],ordered = True,categories=["M", "L", "XL"])
category 
```

```python
# size str dtype category 
df20['size_category'] = df20['size'].astype('category')
df20
```

```python
# size_category category dtype
# SeriesCategory
# Category
df20.dtypes
```

```python
# ()One-hot Encoding () => get_dummies() 
# DataFrame df23
df23 = pd.DataFrame([['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long']],columns=['color','size','sex','length'])
df23
```

```python
# color One-hot Encoding ()
pf = pd.get_dummies(df23['color'])    
pf
```

```python
# get_dummies() One-hot Encoding DataFrame df23   => size  
# get_dummies() category 
df23 = pd.concat([df23,pf],axis=1)
df23
```

```python
# read_csvcsv (usecols) () ()
df24 = pd.read_csv('structured.csv',usecols=['Name','chinese','math','eng'])
df24
```

```python
# fillna() 
df25 = df24.fillna(0)  # 0
df25
```

```python
# 0...
df24['eng'] = df24['eng'].fillna(df24['eng'].median())    # Jay eng 89
df24['math'] = df24['math'].fillna(df24['math'].mean())   # James math 75
df24
```
