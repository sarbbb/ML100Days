```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'Name':['EVA','EVE','EEE'],
                    'math_score':[92,95,90],
                    'eng_score':[93,90,91]})
df2 = pd.DataFrame({'Name':['EVA','EVE','EEE'],
                    'math2_score':[96,98,90],
                    'bio_score':[91,97,91]})
print((df1['math_score'] + df2['math2_score'])/2)
```

```python
df3 = pd.DataFrame([[3,4,5]])
df4 = pd.DataFrame([[7,8,9]])
print(df3 + df4)
print('='*40)
print(df3 + 1)       # 與常數相加
```

```python
# DataFrame的對齊特性很嚴格
df5 = pd.DataFrame([[3,4,5]],columns = ['a','b','c'])  # df5 與 df6 欄位名稱不同，在相加時，沒有的欄位為NaN
df6 = pd.DataFrame([[7,8,9]],columns = ['c','d','e'])  # 任何數和NaN 相加的結果為 NaN
print(df5 + df6)
```

```python
df7 = pd.DataFrame([[2,3,4,5],
                    [5,4,3,2]],columns = ['a','b','c','d'])
print( df7 > 3 )                 # 用條件做出boolean陣列  => 這裡的條件是用整個dataframe來看
print('='*40)
print(df7[df7 > 3])              # 把boolean陣列做為遮罩，篩選dataframe
print('='*40)
print(df7['b'] > 3)              # 用欄位做為篩選的條件，做出boolean陣列
print('='*40)
print(df7[df7['b'] > 3])         # 只把符合欄位條件的record選取出來
```

```python
# dataframe的排序 sort_values()
df8 = pd.DataFrame([['C',3,4],
                    ['B',9,3],
                    ['a',8,4],
                    ['a',7,2]],columns = ['str','a','b'] )
print(df8.sort_values(by= ['str','b']))                      # 依照欄位str排序，接著再依照欄位b排序，默認升序排序
print('='*40)
print(df8.sort_values(by= ['str','b'],ascending = False))    # 依照欄位str排序，接著再依照欄位b排序，可以用參數改為降序排序
```


圖片參考 取自：cupoy

```python
df9 = pd.DataFrame([[2,3,4],
                    [5,1,3],
                    [8,1,9],
                    [8,2,7]])
print(df9.shape)
print('='*40)
print(df9.idxmin())     # idx是指最小值所在的index
print('='*40)
print(df9.idxmax())
print('='*40)
print(df9.sum())
print('='*40)
print(df9.quantile(0.1))
print('='*40)
print(df9.mode())        # 如果該欄位沒有眾數，就會把欄位的值從小排到大排序顯示
print('='*40)
print(df9.median())
print('='*40)
print(df9.describe())
print('='*40)
print(df9.mean())        # default 預設axis=0
print(df9.mean(axis=1))  # default 預設axis=1為計算每個records的mean
```

```python
# DataFrame自訂義函式應用 apply()
df9.apply(lambda x: x**(0.5)*10)
```
