### 1. 根據題目給的 DataFrame 完成下列操作：

```
- 計算每個不同種類 animal 的 age 的平均數
- 計算每個不同種類 animal 的 age 的平均數
- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
```


````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

````python
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
````

````python
# 計算每個不同種類 animal 的 age 的平均數 (忽略NaN)
df.groupby(by=['animal']).agg({'age':np.mean})  # 先聚合，再求平均 (會忽略NaN)，如果希望NaN也要參與計算，要先做前處理，把NaN替代為0
````

````python
df_new = df.fillna(0)                               # 先做資料前處理，把缺失值填0
df_new
````

````python
# 計算每個不同種類 animal 的 age 的平均數 (包含NaN)
df_new.groupby(by=['animal']).agg({'age':np.mean})   # 再做聚合，求平均
````

````python
# 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
df.sort_values(by = ['age','visits'],ascending = [True,False])  # sort_valuse多欄位，不同順序(age升序；visits降序)
````

````python
# 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
replace_dict = {'yes':True,'no':False}                             # 先建立dict寫明替換的條件
new_df = df.replace({'priority':replace_dict})                     # 用replace()替換
print(new_df)
````

### 2. 一個包含三個欄位的 DataFrame，將每個數字減去

* 1) 該欄位的平均數
* 2) 該筆資料平均數

````python
df1 = pd.DataFrame(np.random.random(size=(5, 3)))
df1
````

````python
# 1個DataFrame，將每個數字減去 該欄位的平均數
df1 - df1.mean()
````

````python
# 1個DataFrame，將每個數字減去 該筆資料平均數  
df1.mean(axis = 1)                        # 先求每筆record的平均值
````

````python
# 1個DataFrame，將每個數字減去 該筆資料平均數
df1.sub(df1.mean(axis = 1),axis = 0)      # 後面的axis = 0是指df1的axis，表示用axis =0 的每一個值減去 df1.mean(axis = 1)
````

### 3. 給定一個DataFrame，請問：

1) 哪一比的資料總合最小
2) 哪一欄位的資料總合最小



````python
# importing pandas as pd
import pandas as pd
 
# Creating the dataframe
df2 = pd.DataFrame({"A":[1, 5, 3, 4, 2],
                   "B":[3, 2, 4, 3, 4],
                   "C":[2, 2, 7, 3, 4],
                   "D":[4, 3, 6, 12, 7]},
                   index =["A1", "A2", "A3", "A4", "A5"])
 
# Print the dataframe
df2
````

````python
# 哪一比的資料總合最小
df2.sum( axis=1 ).argmin()
````

````python
# 哪一比的資料總合最小
df2.sum( axis=0 ).argmin()
````
