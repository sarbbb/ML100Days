```python
# array
import numpy as np

a = np.array([20,30,40,50])
b = np.arange(4)              # b = [0,1,2,3]

print(a+b)
print(a*b)
```

```python
# 
c = np.array([2,4,6,8])
print(c-1)
print(c/2)
```

```python
# 
d = np.array([[1,2],
             [3,4],
             [5,6]])
e = np.array([1,1])

print(d+e)
```

```python
# numpy broadcast  python 

f = np.array([10,20,30,40])
print(f+1)

g = [10,20,30,40]
h = []

for i in g:           # python n numpy
    h.append(i+1)

print(h)
```

```python
# 
i = np.array(([1,2],
              [3,4]))
print(i,'\n')
print(i.transpose(),'\n')
print(np.trace(i))
```
