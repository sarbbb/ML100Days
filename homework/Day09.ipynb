```python
import pandas as pd
```

```python
# dataframe
df = pd.DataFrame()
print(df)
```

```python
df1 = pd.DataFrame([[2,3,4],
                   [5,6,7]],index = ['a','b'],columns = ['x','y','z'])
print(df1)
```

```python
df2 = pd.DataFrame({'Name':['EVE','EVA'],
                    'Score':[96,98]})
print(df2)
```

```python
df3 = pd.DataFrame([{'Name':'EVE','Score':96},
                    {'Name':'EVA','Score':99}])
print(df3)
```

```python
print(df2['Name'])
print('='*40)
print(type(df2['Name']))
```

```python
print(df3[['Name','Score']])
print('='*40)
print(type(df3[['Name','Score']]))
```

```python
sales = [('Jones LLC', 150, 200, 50),
         ('Alpha Co', 200, 210, 90),
         ('Blue Inc', 140, 215, 95)]
labels = ['account', 'Jan', 'Feb', 'Mar']
df9 = pd.DataFrame.from_records(sales, columns=labels)
print(df9)
```
