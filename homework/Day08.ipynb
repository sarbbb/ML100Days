```python
import pandas as pd
```

```python
print(pd)
print(pd.__version__)
```

```python
s = pd.Series([1,2,3])
print(s)                   # 左邊(0 1 2)是索引
print('='*40)
print(type(s))
print('='*40)
print(s.shape)
print('='*40)
print(s.size)
print('='*40)
print(s.dtype)
```

```python
df = pd.DataFrame([[1,2,3],[4,5,6]])
print(df)                  # 左邊(0 1)是row的索引，上面(0 1 2)是欄位名稱
print('='*40)
print(type(df))
print('='*40)
print(df.shape)
print('='*40)
print(df.size)
print('='*40)
print(df.dtypes)
```
