```python
# 創建一個ndarray
a = np.arange(9).reshape(3,3)
print(a)
print(type(a))
```

```python
# ndarray 與 type 有關的屬性
print(a.dtype)     # 陣列中每個元素的type
print(a.itemsize)  # 陣列中每個元素佔的記憶體空間
```

```python
# 用astype()修改資料型態
a = a.astype('int64')
print(a.dtype)
print(a.itemsize)
```

```python
# 判斷資料型態
print(a.dtype == 'int64')   # 只判斷資料型態
print(a.dtype is np.int64)  # 用 is 會除了資料型態外，連記憶體空間都要一樣 
```

```python
# 正確的資料型態判斷
print(a.dtype == 'int64')
print(a.dtype == np.int64)
print(a.dtype is np.dtype('int64'))  #用 is 要寫到np的dtype屬性。 因為 is 會比較object是否相同(連記憶體位址都要一樣)
```

```python
# type() 與 np.dtype的差別  => type(a) 是檢查變數的型態，而不是ndarray的元素型態； a.dtype才會回傳ndarray元素的型態
print(type(a))
print(a.dtype)
```
