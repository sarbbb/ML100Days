```python
# 相同大小的array四則運算
import numpy as np

a = np.array([20,30,40,50])
b = np.arange(4)              # b = [0,1,2,3]

print(a+b)
print(a*b)
```

```python
# 常數與陣列做四則運算
c = np.array([2,4,6,8])
print(c-1)
print(c/2)
```

```python
# 不同陣列大小的的則運算
d = np.array([[1,2],
             [3,4],
             [5,6]])
e = np.array([1,1])

print(d+e)
```

```python
# numpy陣列 broadcast 與 python 的比較

f = np.array([10,20,30,40])
print(f+1)

g = [10,20,30,40]
h = []

for i in g:           # 原本的python 要用迴圈走n次執行， numpy用陣列直接計算
    h.append(i+1)

print(h)
```

```python
# 陣列的矩陣運算
i = np.array(([1,2],
              [3,4]))
print(i,'\n')
print(i.transpose(),'\n')
print(np.trace(i))
```
