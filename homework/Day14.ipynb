```python
df20 = pd.DataFrame([['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long']],columns=['color','size','sex','length'])
df20
```

```python
# 用sklearn 的LabelEncoder 實現類別資料(順序性)的編碼   (編碼順序是依ord()順序)
from sklearn.preprocessing import LabelEncoder
class_label = LabelEncoder()                     # 要把LabelEncoder() 賦給class_label，這樣之後才能將label inverse回來
df20['size_label'] = class_label.fit_transform(df20['size'].values)  # 將size 值用LabelEncoder 編順序 (此順序不可自訂，依python ord()順序)
df20
```

```python
# 再用inverse_transform 可把label 還原回原本的值
df20['size_label_inverse'] = class_label.inverse_transform(df20['size_label'])
df20
```

```python
# 因為LabelEncoder() 的編碼是依ord(), 就是依ASCII code, 但這樣的編碼並不符合需求(順序性不符合實際生活經驗)，因此，可以採用map 的方式『正確排序』
df20["size_label_map"]=df20["size"].map({'M': 0, 'L': 1,'XL':2})
df20
```

```python
# 先用Categorical() 創建category => 用category建立可以指定類別與類別順序 (如：此例 ['M' < 'L' < 'XL'])
category = pd.Categorical(df20['size'],ordered = True,categories=["M", "L", "XL"])
category 
```

```python
# 再將size 的str dtype 用category 來替代
df20['size_category'] = df20['size'].astype('category')
df20
```

```python
# 上面新增的size_category 變為category 的dtype
# Series資料類型：Category
# 使用Category資料的一個好處就是：可以很好的節省在時間和空間的消耗。
df20.dtypes
```

```python
# 將類別資料(無順序性)做One-hot Encoding (一位有效編碼) => 用get_dummies() 做
# 創建一個筆數多的DataFrame df23
df23 = pd.DataFrame([['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long'],
                     ['green','M','male','short'],
                     ['red','L','female','normal'],
                     ['blue','XL','male','long']],columns=['color','size','sex','length'])
df23
```

```python
# 指定color 欄位做One-hot Encoding (一位有效編碼)
pf = pd.get_dummies(df23['color'])    
pf
```

```python
# 將上面get_dummies() 做One-hot Encoding 的DataFrame 與原本的df23 拼貼在一起  => 如此，類別size 欄位也可以用於分析 
# 但要注意，使用get_dummies() 會隨著category 越多，資料維度越大，而資料維度越大，分析會有越多問題
df23 = pd.concat([df23,pf],axis=1)
df23
```

```python
# 用read_csv讀取csv檔 (用usecols指定挑選的欄位) (原始檔有欄名) (有缺失值)
df24 = pd.read_csv('structured.csv',usecols=['Name','chinese','math','eng'])
df24
```

```python
# 缺值處理方法，用fillna() 將所有缺失值填補上固定數值
df25 = df24.fillna(0)  # 將缺失值補上0
df25
```

```python
# 缺失值除了補0以外，也可以補中位數、平均數...等
df24['eng'] = df24['eng'].fillna(df24['eng'].median())    # 將Jay 的eng 缺失值填中位數89
df24['math'] = df24['math'].fillna(df24['math'].mean())   # 將James 的math 缺失值填平均值75
df24
```

```python
# 用read_csv讀取csv檔 (用usecols指定挑選的欄位) (原始檔有欄名) (有缺失值)
df26 = pd.read_csv('structured.csv',usecols=['Name','chinese','math','eng'])
df26
```

```python
# 缺失值用『前(後)補值』 => method ='ffill' 是補上一筆的資料； method = 'bfill' 是補後一筆的資料
df27 = df26.fillna(method = 'ffill')  # 這個補值法會有一個問題，就是如果缺失值在第一筆，還是會NaN
df27
```

```python
# 缺失值用『(前)後補值』 
df28 = df26.fillna(method = 'bfill')  # 這個補值法會有一個問題，就是如果缺失值在最後一筆，還是會NaN
df28
```
